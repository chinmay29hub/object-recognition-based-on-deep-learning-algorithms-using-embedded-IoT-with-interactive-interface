```python
import cv2
import numpy as np
from translate import Translator
import pyttsx3
import translators as ts
#loading yolo
c =[]#C IS ARRAY TO STORE DET OBJ
myval=[]#MY val IS ARRAY TO STORE DET OBJ
net = cv2.dnn.readNet('yolov3-tiny.weights', 'yolov3.cfg') #yolo configration and re trained model weights

def chandan(V):
    global c 
    
    classes = []
    with open("coco.names", "r") as f: # this are text of name files that are to be identifies
        classes = f.read().splitlines()

    cap = cv2.VideoCapture(V) # input
    font = cv2.FONT_HERSHEY_PLAIN
    colors = np.random.uniform(0, 255, size=(100, 3))

    while True:
        _, img = cap.read() # we load all frame of the video running in while loop and store each frame using cap.read 

        height, width, _ = img.shape
        # detecting objects 

        blob = cv2.dnn.blobFromImage(img, 1/255, (416, 416), (0,0,0), swapRB=True, crop=False)
        net.setInput(blob)
        output_layers_names = net.getUnconnectedOutLayersNames()
        layerOutputs = net.forward(output_layers_names)


        boxes = []
        confidences = []
        class_ids = []

        #showing info on the screen
        for output in layerOutputs:
            for detection in output:
                scores = detection[5:]
                class_id = np.argmax(scores)
                confidence = scores[class_id]
                if confidence > 0.1:
                    #obj detected 
                    center_x = int(detection[0]*width)
                    center_y = int(detection[1]*height)
                    w = int(detection[2]*width)
                    h = int(detection[3]*height)
                    #rectangle coordinates
                    x = int(center_x - w/2)
                    y = int(center_y - h/2)

                    boxes.append([x, y, w, h]) #show rectangle area
                    confidences.append((float(confidence)))# how much is confidence
                    class_ids.append(class_id)#name of obje that was detected

        indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.1, 0.4)

        if len(indexes)>0:
            for i in indexes.flatten():
                x, y, w, h = boxes[i]
                label = str(classes[class_ids[i]])
                c.append(label)
                confidence = str(round(confidences[i],2))
                color = colors[i]
                cv2.rectangle(img, (x,y), (x+w, y+h), color, 2)
                cv2.putText(img, label + " " + confidence, (x, y+20), font, 2, (255,255,255), 2)
        
        cv2.imshow('Image', img)
        key = cv2.waitKey(1)
        if key==27:
            break

    cap.release()
    cv2.destroyAllWindows()

#-------------------------------------------------------------------------------------------------------------
def chandan2(M):
    classes = []
    with open("coco.names","r") as f:
        classes = [line.strip() for line in f.readlines()]
    layer_names = net.getLayerNames()
    output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]
    colors= np.random.uniform(0,255,size=(len(classes),3))
    #loading image
#     img = cv2.imread("yo1.JPG")
    img = cv2.imread(M)

    #img = cv2.resize(img,None,fx=0.4,fy=0.3
    height,width,channels = img.shape
    #detecting objects
    blob = cv2.dnn.blobFromImage(img,0.00392,(416,416),(0,0,0),True,crop=False)
    # for b in blob:
    #     for n,img_blob in enumerate(b):
    #         cv2.imshow(str(n),img_blob)

    net.setInput(blob)
    outs = net.forward(output_layers)
    #print(outs[1])


    #Showing info on screen/ get confidence score of algorithm in detecting an object in blob
    class_ids=[]
    confidences=[]
    boxes=[]
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.2:
                #onject detected
                center_x= int(detection[0]*width)
                center_y= int(detection[1]*height)
                w = int(detection[2]*width)
                h = int(detection[3]*height)

                cv2.circle(img,(center_x,center_y),10,(0,255,0),2)
                #rectangle co-ordinaters
                x=int(center_x - w/2)
                y=int(center_y - h/2)
                cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

                boxes.append([x,y,w,h]) #put all rectangle areas
                confidences.append(float(confidence)) #how confidence was that object detected and show that percentage
                class_ids.append(class_id) #name of the object tha was detected

    indexes = cv2.dnn.NMSBoxes(boxes,confidences,0.4,0.6)

    global myval
    font = cv2.FONT_HERSHEY_PLAIN
    i =0
    for i in range(len(boxes)):
        if i in indexes:
            x,y,w,h = boxes[i]
            label = str(classes[class_ids[i]])
            color = colors[i]
            myval.append(label)
            cv2.rectangle(img,(x,y),(x+w,y+h),color,2)
            cv2.putText(img,label,(x,y+30),font,1,(255,255,255),2)
    print(myval)
    cv2.imshow("Image",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# ___________________________________NEXT PART_______________________________________________________

def voice(K):
#     k = "person"
#     translator= Translator(to_lang="German")

    
#         translator = Translator(to_lang="German")
    translation = translator.translate(K)
    print(translation)

    # Python program to show
    # how to convert text to speech

    # Initialize the converter
    converter = pyttsx3.init()

    # Set properties before adding
    # Things to say

    # Sets speed percent 
    # Can be more than 100
    converter.setProperty('rate', 150)
    # Set volume 0-1
    converter.setProperty('volume', 0.7)

    # Queue the entered text 
    # There will be a pause between
    # each one like a pause in 
    # a sentence
    converter.say("this is a "+ran+ " and it is called "+translation+"in"+g)
    converter.say("I'm also a geek")

    # Empties the say() queue
    # Program will not continue
    # until all speech is done talking
    converter.runAndWait()
    
option = "Y"
while(option=="Y"):
    
    print("select langauge 1 for eng | 2 for german | 3 for hindi")
    a = int(input()) # taking input from user for selecting langaufe
    if a==1:
        g ="French" # in g we are saving selected lang
    elif a==2:
        g="German"
    else :
        g = "Japanese"
    print(g)
    translator= Translator(to_lang = g)
    # translation = translator.translate("select option 1 for test and 2 for detecting a object")
    # print(translation)
    j ="select option 1 for test and 2 for detecting a object" 
    print("select option 1 for test and 2 for detecting a object")
    # translating
    converter = pyttsx3.init()

    converter.setProperty('rate', 150)
    # Set volume 0-1
    converter.setProperty('volume', 0.7)

    converter.say(j)

    converter.runAndWait()

    n  = int(input())
    ZZ=0#here it is to take camera input 
    if n==1:
        # Python code to pick a random
        # word from a text file
        import random

        # Open the file in read mode
        with open("random val.txt", "r") as file:
            allText = file.read()
            words = list(map(str, allText.split()))
            ran = random.choice(words)
        # print random string
            

        translation = translator.translate(ran)
        print(translation)
        print("bring me a object which is called "+ran+" in "+g)
        chandan(ZZ)
        voice(ran)

        if ran in c:
            voice(ran)
    else:
        cam = cv2.VideoCapture(0)

        cv2.namedWindow("test")

        img_counter = 0

        while True:
            ret, frame = cam.read()
            if not ret:
                print("failed to grab frame")
                break
            cv2.imshow("test", frame)

            k = cv2.waitKey(1)
            if k%256 == 27:
                # ESC pressed
                print("Escape hit, closing...")
                break
            elif k%256 == 32:
                # SPACE pressed
                img_name = "chandan_{}.JPG".format(img_counter)
                cv2.imwrite(img_name, frame)
                print("{} written!".format(img_name))
                img_counter += 1
#                 global img_name

        cam.release()

        cv2.destroyAllWindows()
#         m = "chandan_{}.JPG".format(img_counter)
        m=img_name
        print(m)
        chandan2(m)
        

        for k in myval:
            voice(k)
    option = input("do you want to continue Y/N ")
    
    
```
